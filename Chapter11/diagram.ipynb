```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Function to draw rectangles with optional labels
def draw_rectangle(ax, xy, width, height, label="", color="black", linestyle="-", label_position="below"):
    rect = patches.Rectangle(xy, width, height, linewidth=1, edgecolor=color, facecolor='none', linestyle=linestyle)
    ax.add_patch(rect)
    if label:
        if label_position == "center":
            ax.text(xy[0] + width / 2, xy[1] + height / 2, label, ha="center", va="center", fontsize=10, color=color)
        elif label_position == "below":
            ax.text(xy[0] + width / 2, xy[1] - 0.2, label, ha="center", va="top", fontsize=10, color=color)

# Setup plot
fig, ax = plt.subplots(figsize=(12, 4))  # Increased width for better spacing
ax.set_xlim(0, 12)
ax.set_ylim(0, 5)
ax.axis("off")

# Draw Input Image
draw_rectangle(ax, (0.5, 2), 1, 1, label="Input\nImage", label_position="below")  # "Input Image" in two lines

# Draw Convolution Layer
draw_rectangle(ax, (2, 1.5), 1.5, 2, label="Convolution", label_position="below")
for i in range(5):  # Simulate multiple feature maps
    draw_rectangle(ax, (2 + i*0.15, 1.5 + i*0.15), 1.5, 2, color="gray", linestyle="--")

# Draw Pooling Layer
draw_rectangle(ax, (4, 1.5), 1.5, 2, label="Pooling", label_position="below")
for i in range(5):  # Simulate multiple pooled feature maps
    draw_rectangle(ax, (4 + i*0.15, 1.5 + i*0.15), 1.5, 2, color="gray", linestyle="--")

# Draw Flattening Layer
draw_rectangle(ax, (6, 1.5), 0.5, 2, label="Flattening", label_position="below")

# Draw Fully Connected (Dense) Layer
draw_rectangle(ax, (7, 1.5), 2.5, 2, label="Full Connection\n(Dense)", label_position="below")
for i in range(5):  # Simulate multiple dense layers
    draw_rectangle(ax, (7 + i*0.15, 1.5 + i*0.15), 2.5, 2, color="gray", linestyle="--")

# Draw Output Layer
draw_rectangle(ax, (10.5, 2), 1, 1, label="Output", label_position="below")  # Moved further for spacing

# Draw connections with arrows
# Input to Convolution
ax.annotate("", xy=(2, 2.5), xytext=(1.5, 2.5), arrowprops=dict(arrowstyle="->", color="black"))
# Convolution to Pooling
ax.annotate("", xy=(4, 2.5), xytext=(3.5, 2.5), arrowprops=dict(arrowstyle="->", color="black"))
# Pooling to Flattening
ax.annotate("", xy=(6, 2.5), xytext=(5.5, 2.5), arrowprops=dict(arrowstyle="->", color="black"))
# Flattening to Fully Connected
ax.annotate("", xy=(7, 2.5), xytext=(6.5, 2.5), arrowprops=dict(arrowstyle="->", color="black"))
# Fully Connected to Output
ax.annotate("", xy=(10.5, 2.5), xytext=(9.5, 2.5), arrowprops=dict(arrowstyle="->", color="black"))

plt.show()
```
